```python
import numpy as np
import matplotlib.pyplot as plt

# Load actual data from uploaded file
file_path = "/content/drive/MyDrive/phy264/Time intervals (1).txt"
data = np.loadtxt(file_path)

# Proceed only if data is successfully loaded
if data is not None:
    mu = len(data)/np.sum(data)  # event rate np.a
    #mu = np.average(data)**-1 # event rate

    # Regular histogram setup
    bin_edges_linear = np.linspace(0, 1, 101)  # 100 bins from 0 to 1
    counts, _ = np.histogram(data, bins=bin_edges_linear)  # counts only

    # Bin centers and widths for model histogram
    bin_centers = (bin_edges_linear[:-1] + bin_edges_linear[1:]) / 2
    bin_widths = bin_edges_linear[1:] - bin_edges_linear[:-1]
    N = len(data)

    # Model histogram based on exponential PDF
    pdf_values = mu * np.exp(-mu * bin_centers)
    model_counts = N * pdf_values * bin_widths  # Histogram model

    # Plot regular histogram (experimental) and model
    plt.figure(figsize=(10, 6))
    plt.bar(bin_centers, counts, width=bin_widths, alpha=0.6, label='Experimental Data')
    plt.plot(bin_centers, model_counts, 'r--', linewidth=2, label='Model Histogram')
    plt.title("Regular Histogram vs. Model")
    plt.xlabel("Time between events (s)")
    plt.ylabel("Counts per Bin")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Logarithmic histogram setup
    log_bin_edges = np.logspace(np.log10(1e-7), np.log10(1), 41)  # 10 bins per decade from 100ns to 1s
    counts_log, _ = np.histogram(data, bins=log_bin_edges)

    # Bin centers and widths for log model histogram
    bin_centers_log = (log_bin_edges[:-1] + log_bin_edges[1:]) / 2
    bin_widths_log = log_bin_edges[1:] - log_bin_edges[:-1]

    # Model histogram using log bins
    pdf_values_log = mu * np.exp(-mu * bin_centers_log)
    model_counts_log = N * pdf_values_log * bin_widths_log

    # Plot logarithmic histogram
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(bin_centers_log, counts_log, width=bin_widths_log, alpha=0.6, label='Experimental Data (mz248)')
    ax.plot(bin_centers_log, model_counts_log, 'r--', linewidth=2, label='Model Histogram (mz248)')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_title("Logarithmic Histogram vs. Model")
    ax.set_xlabel("Time between events (s)")
    ax.set_ylabel("Counts per Bin (log scale)")
    ax.legend()
    ax.grid(True, which='both', ls='--')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Load actual data
file_path = "/content/drive/MyDrive/phy264/Station 1. Thursday.txt"
data = np.loadtxt(file_path)

# Proceed only if data is successfully loaded
if data is not None:
    mu = len(data) / np.sum(data)  # Event rate
    N = len(data)

    # --- Long Interval Histogram Setup ---
    bin_edges_short = np.arange(300e-9, 10.1e-6, 100e-9)  # 300ns to 10us, step 100ns
    counts_short, _ = np.histogram(data, bins=bin_edges_short)

    bin_centers_short = (bin_edges_short[:-1] + bin_edges_short[1:]) / 2
    bin_widths_short = bin_edges_short[1:] - bin_edges_short[:-1]

    # --- Model Histogram ---
    pdf_values_short = mu * np.exp(-mu * bin_centers_short)
    model_counts_short = N * pdf_values_short * bin_widths_short

    # --- Curve Fit: PDF(t) = A * exp(-r * t) ---
    def exp_pdf(t, A, r):
        return A * np.exp(-r * t)

    popt, _ = curve_fit(exp_pdf, bin_centers_short, counts_short)

    A_fit, r_fit = popt
    fitted_curve = exp_pdf(bin_centers_short, A_fit, r_fit)

    # --- Plot: Experimental vs Model vs Fit ---
    plt.figure(figsize=(10, 6))
    plt.bar(bin_centers_short, counts_short, width=bin_widths_short, alpha=0.6, label='Experimental Data')
    plt.plot(bin_centers_short, model_counts_short, 'r--', linewidth=2, label='Model Histogram')
    plt.plot(bin_centers_short, fitted_curve, 'g-', linewidth=2, label=f'Fitted Curve: Ae^(-δt)\nA={A_fit:.2e}, δ={r_fit:.2e}')
    plt.title("Histogram for Long Intervals (300ns to 10µs)")
    plt.xlabel("Time between events (s)")
    plt.ylabel("Counts per Bin")
    plt.legend()
    plt.grid(True)
    plt.show()

    # --- Extra Counts Plot ---
    extra_counts = counts_short - model_counts_short
    fit_extra = fitted_curve - model_counts_short

    plt.figure(figsize=(10, 6))
    plt.bar(bin_centers_short, extra_counts, width=bin_widths_short, color='purple', alpha=0.7, label='Extra Counts')
    plt.plot(bin_centers_short, fit_extra, 'g-', linewidth=2, label=f'Fitted Curve: Ae^(-δt)\nA={A_fit:.2e}, δ={r_fit:.2e}')
    plt.title("Histogram of Extra Counts (Experimental - Model)")
    plt.xlabel("Time between events (s)")
    plt.ylabel("Extra Counts per Bin")
    plt.legend()
    plt.grid(True)
    plt.show()
```
