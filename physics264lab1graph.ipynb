```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import re

# Initialize lists to store frequencies and positions
frequencies = []
x_positions = []
y_positions = []

# List of image files and extract frequencies from filenames
image_files = [
    "/content/drive/MyDrive/phy264/lab1data/112Hz.jpg", "/content/drive/MyDrive/phy264/lab1data/128Hz14.jpg", "/content/drive/MyDrive/phy264/lab1data/152Hz18.jpg", "/content/drive/MyDrive/phy264/lab1data/165Hz20.jpg",
    "/content/drive/MyDrive/phy264/lab1data/174Hz22.jpg", "/content/drive/MyDrive/phy264/lab1data/184Hz24.jpg", "/content/drive/MyDrive/phy264/lab1data/192Hz26.jpg", "/content/drive/MyDrive/phy264/lab1data/200Hz28.jpg",
    "/content/drive/MyDrive/phy264/lab1data/207Hz30.jpg", "/content/drive/MyDrive/phy264/lab1data/214Hz32.jpg", "/content/drive/MyDrive/phy264/lab1data/220Hz34.jpg", "/content/drive/MyDrive/phy264/lab1data/230Hz38.jpg",
    "/content/drive/MyDrive/phy264/lab1data/242Hz44.jpg", "/content/drive/MyDrive/phy264/lab1data/249Hz48.jpg", "/content/drive/MyDrive/phy264/lab1data/259Hz56.jpg", "/content/drive/MyDrive/phy264/lab1data/265Hz62.jpg",
    "/content/drive/MyDrive/phy264/lab1data/274Hz74.jpg", "/content/drive/MyDrive/phy264/lab1data/280Hz82.jpg", "/content/drive/MyDrive/phy264/lab1data/285Hz90.jpg", "/content/drive/MyDrive/phy264/lab1data/289Hz94.jpg",
    "/content/drive/MyDrive/phy264/lab1data/300Hz98.jpg", "/content/drive/MyDrive/phy264/lab1data/97Hz10.jpg"
]

# Process each image
for file_name in image_files:
    # Extract frequency from filename using regex
    frequency = int(re.search(r"(\d+)Hz", file_name).group(1))
    frequencies.append(frequency)

    # Load and convert image to grayscale (2D)
    image = Image.open(file_name).convert('L')
    image_data = np.array(image)

    # Find the coordinates of the brightest spot
    brightest_spot = np.unravel_index(np.argmax(image_data, axis=None), image_data.shape)
    y, x = brightest_spot  # (row, column)
    y_positions.append(y)
    x_positions.append(x)

# Convert to numpy arrays for analysis
frequencies = np.array(frequencies)
x_positions = np.array(x_positions)
y_positions = np.array(y_positions)


# Perform polynomial fits of degree 1 (linear fit)
coeff_x = np.polyfit(frequencies, x_positions, 1)  # For x positions
coeff_y = np.polyfit(frequencies, y_positions, 1)  # For y positions

print(coeff_x)



# Generate best-fit lines
fit_x = np.polyval(coeff_x, frequencies)
fit_y = np.polyval(coeff_y, frequencies)

# Plot the horizontal positions with the best-fit line
plt.figure(figsize=(8, 5))
plt.scatter(frequencies, x_positions, label="Measured X Positions", color='green')
plt.plot(frequencies, fit_x, color='orange', label=f"Best Fit: y = {coeff_x[0]:.3f}f + {coeff_x[1]:.3f}")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Horizontal Position (pixels)')
plt.title('Horizontal Position of Brightest Spot vs Frequency')
plt.legend()
plt.grid(True)
plt.show()

# Plot the vertical positions with the best-fit line
plt.figure(figsize=(8, 5))
plt.scatter(frequencies, y_positions, label="Measured Y Positions", color='blue')
plt.plot(frequencies, fit_y, color='red', label=f"Best Fit: y = {coeff_y[0]:.3f}f + {coeff_y[1]:.3f}")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Vertical Position (pixels)')
plt.title('Vertical Position of Brightest Spot vs Frequency')
plt.legend()
plt.grid(True)
plt.show()

# Report fitting parameters
print(f"Fitting Parameters for Horizontal Position (x): A = {coeff_x[0]:.3f}, B = {coeff_x[1]:.3f}")
# print(f"Fitting Parameters for Vertical Position (y): A = {coeff_y[0]:.3f}, B = {coeff_y[1]:.3f}")
```
